# Build your own model
Start with 100 mice, 5% are dark colored and the rest are light colored. Consider the survival rates to be 90% for the dark mice and 80% for the light colored mice. Each surviving mouse reproduces 3 identical offspring at reproductive age and then dies. Estimate the number of generations required for the dark mice population to exceed 95%. How is population size changing over time?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
initial_population = 100
dark_percentage = 0.05
light_percentage = 1 - dark_percentage
dark_survival_rate = 0.9
light_survival_rate = 0.8
offspring_per_mouse = 3

# Initialize population
dark_population = int(initial_population * dark_percentage)
light_population = int(initial_population * light_percentage)

# Track population sizes over time
dark_population_sizes = [dark_population]
light_population_sizes = [light_population]

# Simulate population growth
generations = 0
while dark_population < 0.95 * initial_population:
    # Calculate offspring
    dark_offspring = dark_population * offspring_per_mouse
    light_offspring = light_population * offspring_per_mouse

    # Apply survival rates
    dark_population = int(dark_offspring * dark_survival_rate)
    light_population = int(light_offspring * light_survival_rate)

    # Update population sizes
    dark_population_sizes.append(dark_population)
    light_population_sizes.append(light_population)

    generations += 1

# Plot population sizes over time
plt.plot(range(generations + 1), dark_population_sizes, label='Dark Mice')
plt.plot(range(generations + 1), light_population_sizes, label='Light Mice')
plt.xlabel('Generations')
plt.ylabel('Population Size')
plt.legend()
plt.show()

# Print the number of generations required for the dark mice population to exceed 95%
print(f"Number of generations required: {generations}")
```
