# Ordinary Differential Equations


```python
#imports
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#growth model
def model(t,N,r,K):
    return r*N*(1-N/K)
```

```python
#params
r = 0.5  # Growth rate
K = 100  # Carrying capacity
N0 = 10  # Initial population size
t_span = (0, 100)  # Time span for the simulation
t_eval = np.linspace(*t_span, 1000)  # Time points at which to store the computed solutions
```

```python
#Solve the ODE
sol = solve_ivp(model, t_span, [N0], args=(r, K), t_eval=t_eval, method='RK45')
print(sol)
```

```python
#plots
plt.plot(sol.t, sol.y[0], label='Logistic Growth')
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.title('Logistic Population Growth')
plt.ylim(0,120)
plt.legend()
plt.show()
```
